```python
import pandas as pd
import re
import string
import nltk

from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained('facebook/mbart-large-50-many-to-many-mmt')
```

```python
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt", src_lang= 'en_XX')
```

```python
df = pd.read_csv('Sheet 2-Table 1.csv')
df
```

```python
df.columns =['hindi_sentence','english_sentence']
df
```

```python
df.shape
```

```python
df.english_sentence = df.english_sentence.str.lower()
df
```

```python
def preprocess(text):
  text = ''.join(ch for ch in text if ch not in string.punctuation)
  text = text.lower()
  text = re.sub(r'\d','',text)
  text = re.sub(r'\s+',' ',text)
  text = text.strip()
  return text
```

```python
df.english_sentence = [preprocess(str(en)) for en in df.english_sentence]
df
```

```python
df.hindi_sentence = [preprocess(str(en)) for en in df.hindi_sentence]
df
```

```python
df.drop_duplicates()
df.shape
```

```python
text = df.english_sentence
text
```

```python
references=[]
candidates = []

for i in range(15):
  
  model_inputs = tokenizer(text[i],return_tensors='pt')
  model_inputs
  generated_tokens = model.generate(**model_inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"])
  translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
  print(text[i])
  print(translation[0]+'\n')
  references.append(df.hindi_sentence[i].strip().split())
  candidates.append(translation[0].strip().split())
  translation.clear()
```

```python

score=0
for i in range(len(references)):
  score += nltk.translate.bleu_score.sentence_bleu([references[i]], candidates[i],weights=(1.0, 0, 0, 0))


print("BleU score", score)
```
